```python
from llama_index.llms.groq import Groq # importando o método groq a partir do framework llama
```

```python
llm = Groq(model='llama-3.3-70b-versatile', api_key=chat_api)
response = llm.complete('Qual é a substância do aroma do alecrim')
```

```python
paragrafos = response.text.split("\n\n")
for paragrafos in paragrafos:
  print(paragrafos)
  print()
```
