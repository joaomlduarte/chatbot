```python
import gradio as gr
```

```python
from google.colab import userdata
chat_api = userdata.get('api_chat26')
from groq import Groq
```

```python
# Cria um cliente para receber o modelo LLM
client  =Groq(
    api_key=chat_api
)
```

```python
# Cria uma função chamada assistente especialista em química e física
def especialista_fis_quimica(user_prompt:str) -> str:

  try:
    if user_prompt.strip()=='15':
      return 'Encerrando o Chatbot'

    completion = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[
            {'role':'system','content':'Você é um especialista em química e física'},
            {'role':'user','content':user_prompt},
        ],
            temperature=0,
            max_tokens=1024,
            top_p=1,
            stream=False
    )

    # No gradio devolva a resposta
    return completion.choices[0].message.content
  except Exception as e:
    print("\n Erro no Especialista")
    return f'Erro: {e}'

# Cria a interface com o Gradio
iface = gr.Interface(
    fn=especialista_fis_quimica,
    inputs=gr.Textbox(lines=2, placeholder="Digite sua pergunta sobre Física ou Qímica"),
    outputs=gr.Textbox(lines=10, label='Resposta'),
    title="Especialista em Física e Química",
    description="Digite sua pergunta sobre Física ou Química e o especialista responderá.",
    live = False
)

if __name__ == "__main__":
  iface.launch(debug=True, show_error=True)
```
